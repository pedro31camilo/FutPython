```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/beeswarmTutorial.csv')
df
```

```python
df['Player'] = df['Player'].str.split('\\', expand=True)[0]
df
```

```python
df['per90'] = df['Prog'] / df['90s']
df
```

```python
df = df[df['90s'] >= 6.5]
df['Pos'].unique()
```

```python
df = df[df['Pos'] != "GK"].reset_index()
df['Pos'].unique()
```

```python
background = '#313332'
text_color = '#ffffff'

fig, ax = plt.subplots(figsize=(10,5))
fig.set_facecolor(background)
ax.patch.set_facecolor(background)

plt.rcParams['xtick.color'] = text_color
plt.rcParams['ytick.color'] = text_color

spines = ['top', 'bottom', 'left', 'right']
for x in spines:
    if x in spines:
        ax.spines[x].set_color(background)

sns.swarmplot(data=df, x='per90', color='white', zorder=1)

plt.scatter(x=9.879518, y=0, c='red', s=200, zorder=2)
```

```python
df2 = pd.read_csv('data/beeswarm2.csv')
df2
```

```python
df2['Player'] = df2['Player'].str.split('\\', expand=True)[0]
df2 = df2[df2['90s'] >= 6]
df2 = df2[df2['Pos'] != "GK"].reset_index()
df2
```

```python
df2['Prog'] = df2['Prog'] / df2['90s']
df2['1/3'] = df2['1/3'] / df2['90s']
df2['xA'] = df2['xA'] / df2['90s']
df2['KP'] = df2['KP'] / df2['90s']
df2['PPA'] = df2['PPA'] / df2['90s']
df2
```

```python
metrics = ['Prog', '1/3', 'xA', 'Cmp%', 'KP', 'PPA']
```

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 11))
fig.set_facecolor(background)
ax.patch.set_facecolor(background)

counter = 0
counter2 = 0
met_counter = 0

plt.rcParams['xtick.color'] = text_color
plt.rcParams['ytick.color'] = text_color

for i, ax in zip(df2['Player'], axes.flatten()):
    ax.set_facecolor(background)
    spines = ['top', 'bottom', 'left', 'right']
    for x in spines:
        if x in spines:
            ax.spines[x].set_color(background)
    
    sns.swarmplot(data=df2, x=metrics[met_counter], zorder=1,ax=axes[counter, counter2])
    ax.set_xlabel(f'{metrics[met_counter]} per 90', c=text_color)

    for x in range(len(df2['Player'])):
        if df2['Player'][x] == 'Thiago Alcántara':
            ax.scatter(x=df2[metrics[met_counter]][x],y=0,s=200,c='red',zorder=2)
        if df2['Player'][x] == 'İlkay Gündoğan':
            ax.scatter(x=df2[metrics[met_counter]][x],y=0,s=200,c='white',zorder=2)
        else: continue

    met_counter += 1
    if counter2 == 0:
        counter2 += 1
        continue
    if counter2 == 1:
        counter += 1
        counter2 = 0
```
