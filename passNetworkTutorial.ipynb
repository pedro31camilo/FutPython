```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import Pitch, VerticalPitch
```

```python
df = pd.read_csv('data/valladolidA.csv')
df
```

```python
df = df[df['teamId'] == 'Barcelona']
```

```python
df['passer'] = df['playerId']
df['recipient'] = df['playerId'].shift(-1)

passes = df[df['type'] == 'Pass']
successful_passes = passes[passes['outcome'] == 'Successful']
successful_passes
```

```python
subs = df[df['type'] == 'SubstitutionOff']
subs = subs['minute']
firstSub = subs.min()
firstSub
```

```python
successful_passes = successful_passes[successful_passes['minute'] < firstSub]
successful_passes
```

```python
pas = pd.to_numeric(successful_passes['passer'], downcast='integer')
rec = pd.to_numeric(successful_passes['recipient'], downcast='integer')

successful_passes['passer'] = pas
successful_passes['recipient'] = rec
successful_passes
```

```python
average_positions = successful_passes.groupby('passer').agg({'x': 'mean', 'y': ['mean', 'count']})
average_positions.columns = ['x', 'y', 'count']
average_positions
```

```python
pass_between = successful_passes.groupby(['passer', 'recipient']).id.count().reset_index()
pass_between.rename({'id': 'pass_count'}, inplace=True, axis='columns')
pass_between
```

```python
pass_between = pass_between.merge(average_positions, left_on='passer', right_index=True)
pass_between = pass_between.merge(average_positions, left_on='recipient', right_index=True, suffixes=['', '_end'])
pass_between
```

```python
pass_between = pass_between[pass_between['pass_count'] > 5]
pass_between
```

```python
fig, ax = plt.subplots(figsize=(13, 8.5))
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='white')
pitch.draw(constrained_layout=False, ax=ax, tight_layout=True, figsize=(13, 8.5))

rrows = pitch.arrows(1.2*pass_between.x,.8*pass_between.y,1.2*pass_between.x_end,.8*pass_between.y_end,
                     width = 5, headwidth = 5, color = 'w', ax = ax, zorder = 1, alpha = .5)

nodes = pitch.scatter(1.2*average_positions.x,.8*average_positions.y,
                     s = 300, color = '#d3d3d3', edgecolors = 'black', linewidth = 2.5, alpha = 1, zorder = 1, ax=ax)
```
