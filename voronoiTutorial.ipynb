```python
import pandas as pd
from mplsoccer import Pitch, VerticalPitch
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/voronoi.csv')
barcelona_x = [df['x'][i] for i in range(len(df)) if df['team'][i] == 0]
barcelona_y = [df['y'][i] for i in range(len(df)) if df['team'][i] == 0]
barcelona = list(np.zeros((len(df),), dtype=int))

opp_x = [df['x'][i] for i in range(len(df)) if df['team'][i] == 1]
opp_y = [df['y'][i] for i in range(len(df)) if df['team'][i] == 1]
opp = list(np.ones((len(df, )), dtype=int))
```

```python
points = np.column_stack((df.x, df.y))

fig, ax = plt.subplots(figsize=(12, 6))
fig.set_facecolor('#38383b')
ax.patch.set_facecolor('#38383b')

pitch = VerticalPitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc', half=True)
pitch.draw(figsize=(13, 8), constrained_layout=True, tight_layout=False, ax=ax)

x = df.x
y = df.y

team1, team2 = pitch.voronoi(x, y, df.team)
t1 = pitch.polygon(team1, ax=ax, fc='red', ec='white', lw=3, alpha=.4)
t2 = pitch.polygon(team2, ax=ax, fc='blue', ec='white', lw=3, alpha=.4) 
```
