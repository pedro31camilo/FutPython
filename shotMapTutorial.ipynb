```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mplsoccer import Pitch, VerticalPitch
```

```python
text_color = 'white'
```

```python
data = pd.read_csv('data/shotmaps.csv')
data
```

```python
fig, ax = plt.subplots(figsize=(13, 8.5))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')
pitch = VerticalPitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='white', half=True)
pitch.draw(figsize=(13, 8), constrained_layout=True, tight_layout=False, ax=ax)

plt.gca().invert_yaxis()

for shot in range(len(data['x'])):
    if data['outcome'][shot] == 'Saved':
        plt.scatter(data['x'][shot], data['y'][shot], c='green', s=100, alpha=.75)
    elif data['outcome'][shot] == 'Post':
        plt.scatter(data['x'][shot], data['y'][shot], c='blue', s=100, alpha=.75)
    else:
        plt.scatter(data['x'][shot], data['y'][shot], c='red', s=100, alpha=.75)
```
